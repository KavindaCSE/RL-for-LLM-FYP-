# Exploration Notebook

This notebook is used for exploratory data analysis and visualizations related to the reinforcement learning project.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dummy data
data = pd.read_csv('../data/dummy_data.csv')

# Display the first few rows of the dataset
data.head()
```

```python
# Summary statistics of the dataset
data.describe()
```

```python
# Visualize the distribution of rewards
plt.figure(figsize=(10, 6))
sns.histplot(data['reward'], bins=30, kde=True)
plt.title('Distribution of Rewards')
plt.xlabel('Reward')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the relationship between observations and rewards
plt.figure(figsize=(10, 6))
sns.scatterplot(x='observation', y='reward', data=data)
plt.title('Observations vs Rewards')
plt.xlabel('Observation')
plt.ylabel('Reward')
plt.show()
```
